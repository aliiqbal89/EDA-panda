EDA = Exploratory Data Analysis

```python
# plotting count plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,4))

sns.countplot(x=telecom1["Churn"])

plt.title('Non Churned vs Churned', size=10)
plt.show()
```

```python
# pie chart

lab=telecom1["Churn"].value_counts().keys()

val=telecom1["Churn"].value_counts().values

list_of_tuples=list(zip(lab,val))

per_plot=pd.DataFrame(list_of_tuples,columns=['a','b'])

per_plot.groupby(['a']).sum().plot(kind='pie',y='b',autopct='%1.1f%%',title='Percentage of churned and Non churned user')
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=telecom1.State.sort_values(),hue='Churn',data=telecom1)
```

```python
#Write a function that takes a list, removes duplicates, and returns the result as a tuple.
def list_remove_dupli(n):
  dup = list(set(n))
  return tuple(dup)
test_list = [1, 2, 2, 3, 4, 4, 5]
result = list_remove_dupli(test_list)
print(result)
```

```python
#Given a tuple of values, return a set of unique values greater than 5.

```

```python
value_tuple1 = (1,1,2,3,4,5,6,7,10,15)
result = {value for value in value_tuple1 if value > 5}
print(result)
```

```python
def unique_tuple(my_tuple):
  result1 = {value for value in my_tuple if value >5}
  return result1


my_tuple = (3, 6, 8, 4, 5, 7, 8, 6)

result1 = unique_tuple(my_tuple)
print(result1)
```
